- O desafio é conseguir prever as vendas que vamos ter em determinado período com base nos gastos em anúncios nas 3 grandes redes que a empresa investe: TV, Jornal e Rádio
- TV, Jornal e Rádio estão em milhares de reais
- Vendas estão em milhões

#### Importar a Base de dados

```python
import pandas as pd

vendas_dados = pd.read_csv('advertising.csv')
display(vendas_dados)
```

#### Análise Exploratória
- Visualizar como as informações de cada item estão distribuídas
- Correlação entre cada um dos itens

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(vendas_dados.corr(), annot=True, cmap='Greens')
plt.show()
```

#### Preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
from sklearn.model_selection import train_test_split

y = vendas_dados['Vendas']
x = vendas_dados.drop('Vendas', axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

model_regl = LinearRegression()
model_dt = RandomForestRegressor()

model_regl.fit(x_train, y_train)
model_dt.fit(x_train, y_train)
```

#### Teste da AI e Avaliação do Melhor Modelo

- R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
from sklearn import metrics

# Previsões dos modelos
predict_regl = model_regl.predict(x_test)
predict_dt = model_dt.predict(x_test)

# Comparando os modelos
print('Regressão Linear:', metrics.r2_score(y_test, predict_regl))
print('Árvore de Decisão:', metrics.r2_score(y_test, predict_dt))
```

#### Visualização Gráfica das Previsões

```python
vendas_dados_aux = pd.DataFrame()
vendas_dados_aux['y_test'] = y_test
vendas_dados_aux['Previsoes Arvore Decisao'] = predict_dt
vendas_dados_aux['Previsoes Regressao Linear'] = predict_regl

plt.figure(figsize=(15, 6))
sns.lineplot(data=vendas_dados_aux)
plt.show()
```
